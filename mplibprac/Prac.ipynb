```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-10, 11)
y = x**2
```

```python
f1 = plt.figure(1)
plt.figure(figsize = (30,20))
plotpar1 = {'ls':'dashed', 'marker':'o'}

ax1 = plt.subplot(2,2,1) #multiple axes how ?
ax1.plot(x, y, **plotpar1)

ax3 = plt.subplot(2,2,4)
ax3.plot(x, y)







```
